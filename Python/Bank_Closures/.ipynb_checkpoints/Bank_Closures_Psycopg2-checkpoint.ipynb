```python
def cleandate(olddate):
    elements = olddate.split('-')
    day_orig = int(elements[0])
    if day_orig < 10:
        day = '0' + str(day_orig)
    else:
        day = str(day_orig)

    month_orig = elements[1]
    if month_orig ==  'Jan' : month = '01'
    elif month_orig == 'Feb' : month = '02'
    elif month_orig == 'Mar' : month = '03'
    elif month_orig == 'Apr' : month = '04'
    elif month_orig == 'May' : month = '05'
    elif month_orig == 'Jun' : month = '06'
    elif month_orig == 'Jul' : month = '07'
    elif month_orig == 'Aug' : month = '08'
    elif month_orig == 'Sep' : month = '09'
    elif month_orig == 'Oct' : month = '10'
    elif month_orig == 'Nov' : month = '11'
    elif month_orig == 'Dec' : month = '12'

    year_orig = elements[2]
    year = '20' + year_orig

    newdate = year + '-' + month + '-' + day
    return newdate
```

```python
from sqlalchemy import create_engine

db_string = 'postgresql://postgres:postgres@localhost:5432/Bank_Closures'

engine = create_engine(db_string)
conn = engine.raw_connection()
cur = conn.cursor()
```

```python
df = pd.read_csv('/Users/javierorman/Python/Project 1B/banklist_clean.csv')
df.to_sql('banklist', engine, if_exists='replace', index=False)
```

```python
# open and parse through bank list
fh = open('/Users/javierorman/Python/Project 1B/banklist_clean.csv')
list_lines = fh.read().splitlines()
list_lines.remove('bankname,city,st,cert,acqinst,closedate')

for entry in list_lines:
    olddate = re.findall(',([0-9]*-[A-Za-z]*-[0-9][0-9])$', entry)[0]
    newdate = cleandate(olddate)
    entry = entry.replace(olddate,newdate)
    
# print(list_lines)
```

```python
# convert dates such as 27-Jun-14 to 2014-06-27
cur.execute('''SELECT * FROM banklist''')
rows = cur.fetchall()
for row in rows:
    row_id = row[6]
    olddate = row[5]
    newdate = cleandate(olddate)
    cur.execute('''UPDATE banklist SET closedate = newdate WHERE id == row_id''')
```
