```python
import csv
import psycopg2
```

```python
try:
    conn = psycopg2.connect("dbname='Bank_Closures' user='postgres' host='localhost' password='postgres'")
except:
    print("I am unable to connect to the database")

cur = conn.cursor()
```

```python
cur.execute("""
    DROP TABLE IF EXISTS Bank_List_51820;
    CREATE TABLE Bank_List_51820 (
                id SERIAL NOT NULL PRIMARY KEY,
                bank_name VARCHAR(150),
                city VARCHAR(50),
                state VARCHAR(2),
                cert INT,
                acqinst VARCHAR(100),
                date_of_acquisition DATE
           );
""")
```

```python
## Create temporary table and move data from CSV file into it

cur.execute("""
    DROP TABLE IF EXISTS Bank_Temp;
    CREATE TABLE Bank_Temp(
                bank_name VARCHAR(150),
                city VARCHAR(50),
                state VARCHAR(2),
                cert VARCHAR(10),
                acqinst VARCHAR(100),
                date_of_acquisition VARCHAR(10))
""")

```

```python
with open('banklist.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader) # Skip the header row.
    for row in reader:
        print(row)
        cur.execute(
        "INSERT INTO Bank_Temp (bank_name, city, state, cert, acqinst, date_of_acquisition) VALUES (%s, %s, %s, %s, %s, %s)",
        row
    )
```

```python
# Convert date format
cur.execute('SELECT date_of_acquisition FROM Bank_Temp')
```

```python
def cleandate(olddate):
    elements = olddate.split('-')
    day_orig = int(elements[0])
    if day_orig < 10:
        day = '0' + str(day_orig)
    else:
        day = str(day_orig)

    month_orig = elements[1]
    if month_orig ==  'Jan' : month = '01'
    elif month_orig == 'Feb' : month = '02'
    elif month_orig == 'Mar' : month = '03'
    elif month_orig == 'Apr' : month = '04'
    elif month_orig == 'May' : month = '05'
    elif month_orig == 'Jun' : month = '06'
    elif month_orig == 'Jul' : month = '07'
    elif month_orig == 'Aug' : month = '08'
    elif month_orig == 'Sep' : month = '09'
    elif month_orig == 'Oct' : month = '10'
    elif month_orig == 'Nov' : month = '11'
    elif month_orig == 'Dec' : month = '12'

    year_orig = elements[2]
    year = '20' + year_orig

    newdate = year + '-' + month + '-' + day
    return newdate
```
